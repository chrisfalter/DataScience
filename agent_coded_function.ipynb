```python
import os
from google.colab import userdata
api_key = userdata.get('GEMINI_API_KEY')
os.environ['GEMINI_API_KEY'] = api_key
```

```python
from litellm import completion
from typing import List, Dict


def generate_response(messages: List[Dict]) -> str:
    """Call LLM to get response"""
    response = completion(
        model="gemini/gemini-2.0-flash",
        messages=messages,
        max_tokens=3072
    )
    return response.choices[0].message.content

```

```python
system_role = "You are an expert Python software engineer."
function_specification = input("Specify function requirements:")
messages = [
    {"role": "system", "content": system_role},
    {"role": "user", "content": function_specification}
]

response = generate_response(messages)
print(response)
```

```python
messages.append({"role": "assistant", "content": response})
documentation_request = """
Update the function to include comprehensive documentation. Be sure to include
the following: function description, parameter descriptions, return value description,
example usage, and edge cases
"""
messages.append({"role": "user", "content": documentation_request})
documented_response = generate_response(messages)
print(documented_response)
```

```python
unit_test_request = """
Add test cases using Python’s unittest framework. Tests should cover: Basic
functionality, edge cases, error cases, and various input scenarios.
"""
# prev_response_index = 2
# messages[prev_response_index]["content"] = documented_response
messages.append({"role": "assistant", "content": documented_response})
messages.append({"role": "user", "content": unit_test_request})
final_response = generate_response(messages)
print(final_response)
```
